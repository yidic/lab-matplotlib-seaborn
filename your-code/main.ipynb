#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here

fig2,axes2 = plt.subplots(1,2, figsize = (8,5 ))
fig2.suptitle('x and z plot',fontsize = 20,fontweight = 'bold' ) 
axes2[0].plot(x,y, label = 'y = x*2', c = 'red')
axes2[0].set_title('x',fontweight = 'bold')
axes2[1].plot(x,z, color ='green', label = 'z = x**2')
axes2[1].set_title('z',fontweight = 'bold')

for ax in axes2:
    ax.legend()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
import math

fig3,ax3 = plt.subplots(1, 2, figsize=(10,4))
y1 = x**2
y2 = math.e**x


ax3[0].plot(x, y1)
ax3[0].set_title("Normal scale")
ax3[0].set_xscale("log")

ax3[1].plot(x, y2)
ax3[1].set_yscale("log") 
ax3[1].set_title("Logarithmic scale (y)");

for ax in axes2:
    ax.legend()
```
